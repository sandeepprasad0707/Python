```python
################################################################################################################################

# Accessing Characters in Strings by Index in Python

# String Indexing Basics
# Strings in Python are sequences of characters, and each character has a position (index) starting from 0.

# Basic string indexing
text = "Hello"
print(text[0])  # Output: 'H'
print(text[1])  # Output: 'e'
print(text[2])  # Output: 'l'
print(text[3])  # Output: 'l'
print(text[4])  # Output: 'o'

# Trying to access index out of range
# print(text[5])  # Raises IndexError
```

```python
# Negative Indexing
# Python supports negative indexing, where -1 refers to the last character.

text = "Python"
print(text[-1])  # Output: 'n' (last character)
print(text[-2])  # Output: 'o'
print(text[-3])  # Output: 'h'
print(text[-6])  # Output: 'P' (first character)

```

```python
# Slicing Strings
# Slicing allows you to extract substrings using [start:end:step].

text = "Programming"

# Basic slicing
print(text[0:4])    # Output: 'Prog'
print(text[4:])     # Output: 'ramming' (from index 4 to end)
print(text[:4])     # Output: 'Prog' (from start to index 3)
print(text[2:8])    # Output: 'ogramm'
print(text[0:11])

# Using step
print(text[::2])    # Output: 'Pormig' (every 2nd character)
print(text[1::2])   # Output: 'rga'n (every 2nd character starting from index 1)
print(text[::-1])   # Output: 'gnimmargorP' (reverse the string)

# Slicing with negative indices
print(text[-4:])    # Output: 'ming' (last 4 characters)
print(text[-6:-2])  # Output: 'ammi'

```

```python
# Working with String Methods and Indexing

# # Finding indices of characters
sentence = "The quick brown fox jumps over the lazy dog zzz"


# Find first occurrence
print(sentence.index('z'))     # Output: 4
print(sentence.find('z'))      # Output: 40 (returns -1 if not found)

 
```

```python
################################################################################################################
# Finding Index in Python:
# 1. Basic String Index Finding Methods
# The find() Method
# The find() method returns the lowest index where a substring is found, or -1 if not found.

# Basic syntax: string.find(substring, start, end)

# Basic usage
text = "Hello World"
print(text.find("World"))     # Output: 6
print(text.find("o"))         # Output: 4 (first occurrence)
print(text.find("xyz"))       # Output: -1 (not found)

# With start parameter
print(text.find("o", 5))      # Output: 7 (search from index 5 onward)
print(text.find("o", 0, 5))   # Output: 4 (search between indices 0-4)

# Detailed parameter exploration
text = "abracadabra"

# Start parameter - where to begin searching
print(text.find("a", 0))      # Output: 0 (search from beginning)
print(text.find("a", 1))      # Output: 3 (skip first 'a')
print(text.find("a", 4))      # Output: 5
print(text.find("a", 6))      # Output: 7
print(text.find("a", 8))      # Output: 10

# End parameter - where to stop searching (exclusive)
print(text.find("a", 0, 5))   # Output: 0 (search indices 0-4)
print(text.find("a", 2, 8))   # Output: 3 (search indices 2-7)
print(text.find("a", 5, 10))  # Output: 5 (search indices 5-9)


# Negative indices (converted to positive internally)
print(text.find("a", -5))     # Output: 7 (equivalent to text.find("a", 6))
print(text.find("a", -3, -1)) # Output: 10? Let's check


numb=1478552
print(str(numb).find('5'))  # Output: 4 (first occurrence of '5')
```

```python
# The index() Method
# Similar to find() but raises ValueError when substring is not found.

# Basic syntax: string.index(substring, start, end)

text = "Python Programming"

print(text.index("Pro"))      # Output: 7
print(text.index("o"))        # Output: 4
print(text.index("o", 5))     # Output: 9

# When to use index() vs find()
# Use index() when you're sure the substring exists
# Use find() when it might not exist
```
