```python
#############################################################################################################################################
#String Concatenation
# 1. Concatenation Using the + Operator (Basic Method)

first_name = "John"
last_name = " Doe"

full_name = first_name + last_name

print(full_name)
print(first_name,last_name)

#Adding a Space Between Strings

first_name = "John"
last_name = "Doe"

full_name = first_name + " " + last_name #The space " " is treated as another string and gets concatenated between the two names.

print(full_name)
```

```python
#2.Concatenating Multiple Strings

city = "New"
state = "York"
country = "USA"

address = city + ", " + state + ", " + country

print(address)
```

```python
#3.Concatenation Using +=
#The += operator appends one string to another.
message = "Hello"
message += " World"


print(message)

message = "Hello"
message += " World"
message += " Welcome"

print(message)
```

```python
#4.Concatenating Strings with Variables

name = "Sandeep"

greeting = " Welcome " + name

post = "To Python learning program" + greeting 

print(greeting)
print(post)
```

```python
#5. Concatenating String and Number
# # #Note Python cannot directly concatenate strings and integers.
age = 25

message = "Age is " + str(age)

print(message)

name = " Sandeep"

message = str(20)  + name

print(message)
```

```python
#6.Using join() (Recommended for Multiple Strings)
#  " " is the separator 
# join() places a space between each item.
words = ["Python","is","awesome"]

sentence = " ".join(words)

print(sentence)
```

```python
# 7. Using join() with Comma Separator
items = ["Apple", "Banana", "Orange"]

result = ", ".join(items)

print(result)
```

```python
#8.Using join() with Empty Separator
letters = ["P", "Y", "T", "H", "O", "N"]

word = "".join(letters)

print(word)
```

```python
#9. Using f-Strings (Modern Python Method)
# Variables are placed inside {}.
# Python automatically converts values to strings.
name = "John"
age = 25

message = f"My name'sss is {name} and I am {age} years old."
message1 = f'My name is {name} and I am {age} years old.'
print (f"My name'sss is {name} and I am {age} years old.")
print (f'My names is {name} and I am {age} years old.')
print(message)
print(message1)

num1=20
num2=30
add=num1+num2
print(f"The addition of {num1} and {num2} is= {add}'") # alwase prefer this
print(f'The addition of {num1} and {num2} is= {add}') # alwase prefer this
```

```python
#10. f-Strings with Expressions
# Python evaluates the expression inside {}.
a = 10
b = 20

print(f"Sum = {a + b}")
```

```python
#11. Using format() Method
name = "John"
age = 25

message = "My name is {} and I am {} years old".format(name, age)
print("My name is {} and I am {} years old".format(name, age))
print(message)

# Using Named Placeholders
name = "Jose"
age = 28

message = "Name: {n}, Age: {a}".format(n="John", a=25)

print(message)

```
