```python
#########################################################################################################################
# 1. Basic Type Casting Functions
# 1.1 int() - Converting to Integer
# Parameter: x (number or string), base (optional, default 10)

# Arguments & Limitations:

# Strings must contain valid integer literals
# Floats are truncated (not rounded)
# Cannot convert complex numbers
# Cannot convert non-numeric strings

# Basic Examples
print(int(3.14))        # 3 (truncates, not rounds)
print(int(True))        # 1
print(int(False))       # 0

# String Examples
print(int("42"))        # 42
print(int("-17"))       # -17
print(int("  100  "))   # 100 (strips whitespace)

# Different Bases
print(int("1A", 16))    # 26 (hexadecimal)
print(int("1010", 2))   # 10 (binary)
print(int("77", 8))     # 63 (octal)

# Limitations
# int("3.14")           # ValueError: invalid literal
# int("12a")            # ValueError
# int("0x1A")           # ValueError (use base=16 explicitly)


num = "100"
result = int(num)
print(result)
print(type(result))


salary = 45678.99
result = int(salary)
print(result)


print(int(True))
print(int(False))
```

```python
# 1.2 float() - Converting to Float
# Parameter: x (number or string)

# Arguments & Limitations:

# Strings must be numeric or NaN/infinity
# Cannot convert complex numbers
# Precision limitations (floating-point representation)

# Basic Examples
print(float(10))            # 10.0
print(float(True))          # 1.0
print(float("3.14159"))     # 3.14159
print(float("inf"))         # inf
print(float("NaN"))         # nan

# Scientific Notation
print(float("1.2e-4"))      # 0.00012
print(float("3E+6"))        # 3000000.0

# Limitations
# float("12a")              # ValueError
# float(None)               # TypeError
# float("")                 # ValueError


price = "99.99"

result = float(price)

print(result)

age = 25

result = float(age)

print(result)
```

```python
# 1.3 str() - Converting to String
# Parameter: object

# Arguments & Limitations:

# Works with any Python object
# Uses __str__() method of objects
# Not for binary/bytes conversion (use bytes() for that)

# Basic Examples
print(str(123))             # "123"
print(str(3.14))            # "3.14"
print(str(True))            # "True"
print(str(None))            # "None"

# Collections
print(str([1, 2, 3]))       # "[1, 2, 3]"
print(str({"a": 1}))        # "{'a': 1}"

roll_no = 101

result = str(roll_no)

print(result)


pi = 3.14159

result = str(pi)

print(result)
print(type(pi))
print(type(result))
```

```python
# 2. Collection Type Casting
# 2.1 list() - Converting to List
# Parameter: iterable (optional)

# Arguments & Limitations:

# Iterable must support iteration
# Maintains order of iterable
# Can convert strings, tuples, sets, dictionaries (keys only)

# Examples
print(list("hello"))            # ['h', 'e', 'l', 'l', 'o']
print(list("Hello World"))
print(list((1, 2, 3)))          # [1, 2, 3]
print(list({1, 2, 3}))          # [1, 2, 3] (order not guaranteed)
print(list({"a":1, "b":2}))     # ['a', 'b'] (keys only)
print(list(range(5)))           # [0, 1, 2, 3, 4]

# Converting dict items
print(list({"x":1, "y":2}.items()))  # [('x', 1), ('y', 2)]

# Limitations
# list(123)                     # TypeError: 'int' is not iterable



data = (10, 20, 30)
result = list(data)
print(result)
```

```python
# 2.2 tuple() - Converting to Tuple
# Parameter: iterable (optional)

# Arguments & Limitations:

# Creates immutable sequence
# Similar to list() but returns tuple
# More memory efficient than list

# Examples
print(tuple([1, 2, 3]))         # (1, 2, 3)
print(tuple("abc"))             # ('a', 'b', 'c')
print(tuple(range(3)))          # (0, 1, 2)

# Limitations
# tuple(123)                    # TypeError
# tuple() without arg returns empty tuple: ()
```

```python
# 2.3 set() - Converting to Set
# Parameter: iterable (optional)

# Arguments & Limitations:

# Removes duplicates automatically

# No guaranteed order

# Elements must be hashable (immutable)

# Examples
print(set([1, 2, 2, 3, 3]))    # {1, 2, 3}
print(set("hello"))             # {'h', 'e', 'l', 'o'} (removes duplicate 'l')
print(set((1, 2, 1, 3)))        # {1, 2, 3}

# Limitations
# set([1, 2, [3, 4]])          # TypeError: unhashable type: 'list'
# set({"a":1, "b":2})          # {'a', 'b'} (keys only, not values)
```

```python
# 2.4 dict() - Converting to Dictionary
# Parameter: mapping or iterable of key-value pairs, or keyword arguments

# Arguments & Limitations:

# Iterable must contain pairs (key, value)

# Keys must be hashable

# Later keys override earlier ones

# From list of tuples
print(dict([("a", 1), ("b", 2)]))      # {'a': 1, 'b': 2}
print(dict([(1, "one"), (2, "two")]))  # {1: 'one', 2: 'two'}

# From keyword arguments
print(dict(a=1, b=2))                   # {'a': 1, 'b': 2}
print(dict(zip(["x", "y"], [10, 20])))  # {'x': 10, 'y': 20}

# From dictionary
original = {"name": "Alice", "age": 30}
print(dict(original))                   # {'name': 'Alice', 'age': 30}

# Limitations
# dict([(1,2,3)])                     # ValueError: dictionary update sequence element has length 3; 2 is required
# dict([1,2])                          # TypeError: cannot convert dictionary update sequence element #0 to a sequence
```

```python
# 3 complex() - Complex Numbers
# Parameter: real (number or string), imag (optional number)

# Arguments & Limitations:

# Can parse "a+bj" format strings
# imag only used when real is numeric

# Examples
print(complex(3, 4))                # (3+4j)
print(complex(2.5))                 # (2.5+0j)
print(complex("3+4j"))              # (3+4j)
print(complex("2.5"))               # (2.5+0j)

# Arithmetic with complex
c1 = complex(1, 2)
c2 = complex(3, 4)
print(c1 + c2)                      # (4+6j)
print(c1 * c2)                      # (-5+10j)

# Limitations
# complex("3+4")                    # ValueError: complex() arg is a malformed string
# complex("i")                      # ValueError (use j notation)
```

```python
# 3.3 bool() - Boolean Conversion
# Parameter: x (any type, optional)

# Arguments & Limitations:

# Empty/zero values → False
# Non-empty/non-zero → True
# False values: None, False, 0, 0.0, 0j, "", [], {}, set(), range(0)

# False examples
print(bool(0))          # False
print(bool(0.0))        # False
print(bool(0j))         # False
print(bool(""))         # False
print(bool([]))         # False
print(bool({}))         # False
print(bool(None))       # False

# True examples
print(bool(1))          # True
print(bool(-5))         # True
print(bool(" "))        # True (space is not empty)
print(bool([1]))        # True
print(bool({"a":1}))    # True


value = "Hello"
print(bool(value))

value = ""
print(bool(value))

bool("False")
print(bool("False"))  # True, because non-empty string is True



print(bool(1))
print(bool(100))
print(bool(0))
```
